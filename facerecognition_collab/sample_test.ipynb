```python
import os
import numpy as np
import face_recognition

KNOWN_EMB_DIR = "recognised/team_members"
TEST_IMAGE_PATH = "test.jpg"

known_encodings = []
known_names = []

# Load stored embeddings
for file in os.listdir(KNOWN_EMB_DIR):
    if file.endswith(".npy"):
        path = os.path.join(KNOWN_EMB_DIR, file)
        data = np.load(path, allow_pickle=True)

        # Handle both 1D (128,) and 2D (n,128) arrays
        if len(data.shape) == 1:
            known_encodings.append(data)
            known_names.append(os.path.splitext(file)[0])
        else:
            for enc in data:
                known_encodings.append(enc)
                known_names.append(os.path.splitext(file)[0])

# Convert to numpy arrays
known_encodings = np.array(known_encodings)
known_names = np.array(known_names)
# known_encodings = np.ascontiguousarray(known_encodings, dtype=np.float32)

print("Loaded", known_encodings.shape[0], "known embeddings.")

# Load test image
test_image = face_recognition.load_image_file(TEST_IMAGE_PATH)
test_encodings = face_recognition.face_encodings(test_image)

if len(test_encodings) == 0:
    print("❌ No face found in test image!")
else:
    test_encoding = test_encodings[0]

    if known_encodings.shape[0] == 0:
        print("⚠️ No known encodings loaded. Check your recognised folder.")
    else:
        distances = face_recognition.face_distance(known_encodings, test_encoding)
        # enc_matrix = np.vstack(known_encodings)  # shape: (N,128)
        # diff = enc_matrix - test_encoding
        # distances = np.sqrt(np.sum(np.square(diff), axis=1))

        best_match_index = np.argmin(distances)
        best_distance = distances[best_match_index]
        best_name = known_names[best_match_index]

        if best_distance < 0.6:
            print(f"✅ Match found: {best_name} (Distance: {best_distance:.3f})")
        else:
            print(f"❌ No match found. Closest: {best_name} (Distance: {best_distance:.3f})")
```
